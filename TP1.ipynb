```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('/content/sample.csv')
df
```

```python
import nltk
nltk.download('stopwords')
selected_column = df['text']
x=selected_column[0]
```

```python
import re
from nltk.corpus import stopwords

word_filter = r'\b[a-z][a-z]*[a-z]\b'

filtered_texts = []

for text in selected_column:
    text = text.lower()

    words = re.findall(word_filter, text)

    filtered_words = [word for word in words if word not in stopwords.words('english')]

    filtered_texts.append(filtered_words)

print(filtered_texts)
```

```python
filtered_texts[0]
```

```python
from nltk.stem import PorterStemmer
import spacy

# Charger le modèle spaCy en anglais
nlp = spacy.load("en_core_web_sm")


# Lemmatisation avec spaCy
lemmatized_words = [word.lemma_ for word in nlp(' '.join(filtered_texts[0]))]

# Stemming avec NLTK
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_texts[0]]

# Lemmatisation avec spaCy sur tous les textes
all_lemmatized_words = [word.lemma_ for text in filtered_texts for word in nlp(' '.join(text))]

# Stemming avec NLTK sur tous les textes
all_stemmed_words = [stemmer.stem(word) for word in filtered_texts[0]]

# Afficher les résultats
print("Lemmatisé (spaCy):", lemmatized_words)
print("Stemmé (NLTK):", stemmed_words)
print("Lemmatisé sur tous les textes (spaCy):", ', '.join(all_lemmatized_words))
print("Stemmé sur tous les textes (NLTK):", ', '.join(all_stemmed_words))
```
